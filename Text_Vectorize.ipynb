# Preprocessing

```python
import gensim
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import string

# Fonction de prétraitement pour la lemmatisation et la suppression des stopwords
def preprocess(text):
    # Suppression de la ponctuation
    text = text.translate(str.maketrans("", "", string.punctuation))

    # Tokenisation
    tokens = gensim.utils.simple_preprocess(text)

    # Suppression des stopwords
    stop_words = set(stopwords.words("english"))
    tokens = [token for token in tokens if not token in stop_words]

    # Lemmatisation
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token) for token in tokens]

    return tokens
```

```python
preprocessed_data = df["description"].apply(preprocess)
preprocessed_data
```

# Vectorisation des descriptions (CountVectorizer et TfidfVectorizer)

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

# Créer un vecteur de comptage de mots (countvectorizer)
count_vectorizer = CountVectorizer()

# Créer un vecteur de comptage de mots (tfidvectorizer)
tfid_vectorizer = TfidfVectorizer()

# Appliquer le vecteur de comptage de mots à la série de textes prétraités
count_bow  = count_vectorizer.fit_transform(preprocessed_data.apply(lambda x: " ".join(x)))
tfid_bow = tfid_vectorizer.fit_transform(preprocessed_data.apply(lambda x: " ".join(x)))
```

# Vectorisation des descriptions (Doc2vec)

```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument

# Prétraitement des données : séparer chaque phrase en mots individuels et créer des documents tagués
tagged_data = [TaggedDocument(words=sentence.split(), tags=[str(i)]) for i, sentence in enumerate(df["description"])]

# Construction du modèle Doc2Vec
model = Doc2Vec(tagged_data, vector_size=100, window=5, min_count=1, workers=4)

# Vectorisation des phrases
vectors = [model.infer_vector(sentence.split()) for sentence in df["description"]]
vectors
```
